```python
import pandas as pd
# load data file
df = pd.read_csv("data.csv")
# reshape the d dataframe suitable for statsmodels package 

# generate a boxplot to see the data distribution by treatments. Using boxplot, we can 
# easily detect the differences between different treatments
import matplotlib.pyplot as plt
import seaborn as sns
ax = sns.swarmplot(x="algorithm", y="calls", data=df, color='#7d0013')
plt.show()
```

```python
df = pd.read_csv("data.csv")
df_S1=df[df.algorithm.str.contains("-S1")==True]
df_S2=df[df.algorithm.str.contains("-S2")==True]
df_S3=df[df.algorithm.str.contains("-S3")==True]
df_S3_2=df[df.algorithm.str.contains("-S3/2")==True]
df_S2_3=df[df.algorithm.str.contains("-S2/3")==True]
```

```python
calls=[]
backtracks=[]
pvals_call=[]
pvals_backtrack=[]
from scipy import stats
list_of_df=[df_S1,df_S2,df_S3,df_S3_2,df_S2_3]
for i in range(5):
  my_data = (list_of_df[i]["calls"])
  my_data2 =(list_of_df[i]["backtracks"])
  x=stats.shapiro(my_data).statistic
  y=stats.shapiro(my_data2).statistic
  z=stats.shapiro(my_data).pvalue
  t=stats.shapiro(my_data2).pvalue
  pvals_call.append(z)
  pvals_backtrack.append(t)
  calls.append(x)
  backtracks.append(y)
```

```python
df_stats = pd.DataFrame({"Call":calls,"Backtrack":backtracks})
df_stats.index=["S1","S2","S3","S3_2","S2_3"]
df_stats
```
